```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('AirPassengers.csv')
df['Month'] = pd.to_datetime(df['Month']) 
```

```python
df.set_index('Month', inplace=True)
```

```python
df['12_ma'] = df['Passengers'].rolling(window=12).mean()
print(df.head(15))
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Passengers'], label='Actual')
plt.plot(df['12_ma'], color='red', label='12 month avg')
plt.title('Air Passenger Trend Analysis')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['Month_name'] = df.index.month_name()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10,5))
sns.boxplot(data=df, order=months, x='Month_name', y='Passengers', legend=False)
plt.title('Seasonal Box Plot')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
